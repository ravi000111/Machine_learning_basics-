```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

```python
oecd_bli = pd.read_csv("oecd_bli_2015.csv", thousands=',')
gdp_per_capita = pd.read_csv("gdp_per_capita.csv",thousands=',',delimiter='\t',
 encoding='latin1', na_values="n/a")
```

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    # Filter for Life satisfaction in the oecd_bli
    oecd_bli = oecd_bli[oecd_bli["Indicator"] == "Life satisfaction"]
    oecd_bli = oecd_bli[["Country", "Value"]].set_index("Country")
    oecd_bli.columns = ["Life satisfaction"]

    # Prepare GDP per capita
    gdp_per_capita = gdp_per_capita.set_index("Country")[["2015"]]
    gdp_per_capita.columns = ["GDP per capita"]

    # Merge the data
    full_country_stats = pd.merge(oecd_bli, gdp_per_capita, left_index=True, right_index=True)
    
    return full_country_stats

# Now get the data
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]
```

```python
# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()
```

```python
import sklearn.neighbors
model = sklearn.neighbors.KNeighborsRegressor(
 n_neighbors=3)
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make a prediction for Cyprus
X_new = [[22587]] # Cyprus's GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]
```
